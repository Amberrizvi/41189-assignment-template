## 2.4 Discuss how prevailing models/frames are preventing the problem from being fully addressed or solved

The COVID-19 pandemic is exacerbating the gender inequality experienced by women and there are numerous prevailing models and frames which prevent this shadow pandemic from being resolved or even addressed. Policy makers, media and official authorities have shifted the focus from all other social and environmental issues that are taking place and drawing all the attention towards the global pandemic (Fabrizio et al., 2020). The notion that governments need to focus on the immediate problems to ensure the stability of the society rather than solving the anticipated problems of the future is hindering the possibility of even addressing this issue (Colins College, 2017). Policy makers are more focused towards creating solutions to contain and regulate the spread of the virus and are refusing to focus on aiding women who have been disproportionately affected by the COVID-19 outbreak (Fabrizio et al., 2020).
 
Additionally, stereotypes have significantly exacerbated the socially prescribed care roles of women (Alleman & Rumble, 2020). As detailed in section 2.2, social distancing guidelines have forced children to transition into an online learning environment, thus, women are pressured into leaving or postponing their careers to prioritize the education of their children (WGEA, 2020). Moreover, women who have been the frontline responders during this pandemic are unfairly limited to the household and burdened to take on all care responsibilities including the maintaining of hygiene and sanitisation (Wamera, 2020).
 
In the text ‘The Great Lever’, the historian Walter Scheidel details the notion that a catastrophic event such as a plague are forces that will decrease inequality through a process of leveling. By studying the patterns that have occurred in history, Scheidel explains that “inequality never dies peacefully” (Scheidel, 2017). The historian exemplifies this idea through the study of The Black Plague and concluded that it led to the erosion of status barriers and brought about a leveling in society (Scheidel, 2017). The elite ranks shrank in fortune and size and the poorer were given opportunities to earn better pay (Scheidel, 2017). One can draw parallels between this epidemic with the COVID-19 pandemic and believe that the virus will bring about a ‘levelling’ and counteract the inequality women are facing. However, this will not be the case as disregarding gender inequality can have significant consequences on Australia’s economy and lower the GDP (WGEA, 2020).
 
Another model that has seriously impeded the issue of gender inequality from being resolved is the male breadwinner model which has been heavily institutionalized and reinforced in Australia’s policies and culture (Broomhill & Sharp, 2005). This is greatly evident in women’s access to unemployment benefits as “under the breadwinner model, the institutional structure reinforces the supposition that a woman will be, at best, a secondary earner and so her entitlement may be conditional upon either her husband’s labour force status and/or his income” (Mitchell, n.d.). The male bead winner model is a gender division ideal in which the man is given the responsibility of earning an income to support his family while the women take on domestic labour and care for the family (Parliament of Australia, 2016). The parliament of Australia states that “policy development is still premised on the assumption of the male breadwinner model” (Parliament of Australia, 2020). Consequently, this model creates a power imbalance in the household where men assume the power, this imbalance can create an unhealthy stress in the relationship leading to marital conflicts. Thus, when marital conflicts arise a high level of negative interactions occur including domestic violence which positions women in a vulnerable situation (Cerrato & Cifre, 2018).

## 3.1 Based on your self-study of the supermarket of models, explain the three models that you have chosen to look at the problem

*Markov Models*

The Markov model is a model that determines the future state of sequences by considering the previous states. It holds the assumptions that there are a limited number of states. It is not a simple cycle, since eventually any state can get from one to another.
Due to the fact that inequality is an issue, especially for women, the COVID-19 situation made it worse for them. As mentioned earlier in section 2.1, women are already affected in the job market negatively, and the pandemic would have made it worse. The Markov model is a good way to predict future events, while taking the previous events in the same sequence into account. This model shows that future states change according to previous states. Therefore, this is a good method to predict the impact of COVID-19 on women in certain areas.


*Preferential Attachment Models*

Preferential attachment models reflect circumstances in which nodes are biased towards other nodes that they connect with. This results in a network in which those with more nodes will continually receive more connections (Wang, 2020). This model was chosen due to its ability to emulate unequal power distributions in models. In the circumstance of women’s employment, this can be applied to highlight the prepetual effect of discrimination and disadvantage once some degree of inequality is already in effect. 

*System Dynamic models*

System dynamic models use the probability of several influencing factors to determine the rate of increase/decrease over time. It can be used to predict outcomes when one or more factors within a system are altered.

*Preferential Attachment Models*

Preferential attachment models work by taking the idea that an entity entering the network will have some probability of creating a new form, and some probability of connecting with an existing entity. As this probability accumulates, new nodes or entities joining will then prefer to connect with entities that already have connections, creating a preferential attachment. This can be applied to examine the effect on women-dominated industries in the economy when government investment and stimulus is distributed unequally, favouring male-dominated industries. Using preferential attachment models to investigate this issue can demonstrate its long-term effects on inequality in women’s employment.
