# 1. Introduction

## 1.1 State the problem, why it is important/relevant to you and/or more broadly to society

While COVID-19 has directly and indirectly impacted a wide range of individuals, there has been a disproportionate negative effect on women and gender inequality in Australia due to the government, industry and societal response the pandemic has warranted. This issue is significant as it has the potential to stall and regress progressions made by women in the workplace, the domestic sphere and the political environment. Specifically, the pandemic exasperates the heightened consequences women face due to their over-representation in the healthcare system, financial instability in sectors where they have a larger proportion of employment, increased domestic responsibilities and greater risk of violence and abuse against women (Madgavkar et al, 2020). In addition to these personal implications, the issue is relevant to wider society due to the monetary impacts of a smaller and less diverse workforce created as a result of disruptions to women’s employment (WGEA, 2020).

**Figure 2.1.1**

Figure 2.1.1 summarises how the employment of women in Australia has had a direct effect on gender inequality following COVID-19. 
- Firstly, a significant amount of front-line workers, specifically 75% of health professionals and 88% of midwifery and nursing professionals are women (WGEA, 2020). This has placed a disproportionately large amount of risk of contracting the virus and as a result increased stress due to the pressure on this industry. 
- In addition, sectors with the largest reduction in payrolls from March to April, accommodation and food services, and the second largest, being arts and recreation, both employ significantly more women than men (Crabb, 2020).
- The slightly larger proportion of women in casual employment translates to a larger likelihood of more women losing jobs or decreasing their hours worked.
- Of the industries with the greatest reductions in payrolls, as mentioned prior, majority of their short term ( <12 months) casual workers are women (Cassells & Duncan, 2020). As the ‘JobKeeper’ package initially encompassed those employed for 12 months from March 1st 2020 (Australian Taxation Office, 2020), some of those hardest hit from the exclusion parameters have been and will continue to be women.
- The childcare sector lost access to ‘JobKeeper’ as early as July 20 2020, although all other industries received these payments until September 27. This was justified by the government due to a subsidy which allowed Australian parents free childcare concluding on July 13 2020 (Stewart, 2020).
- As a result of this, the government initiative which allowed Australians to withdraw from their superannuation during the pandemic was used by more women than men (AMP, 2020) (Batchelor, 2020). This has provided necessary support for women during the pandemic, however it has also widened the gender superannuation gap by 4%, with men having an average of 29% higher superannuation balance. 
All of these factors will likely have regressive long-term effects on advances that have made for equity and will likely lower global gross domestic product (GDP), as summarised by the ‘larger impacts’ section of Figure 2.1.1. 

**Figure 2.2.1**

Figure 2.1.2 demonstrates how domestic responsibilities placed on women during the pandemic have widened the gender inequality gap.
- Women already perform almost one third more of their normal working hours on additional unpaid work than men (WGEA, 2020), which is likely to increase during the COVID-19 pandemic due to social gender role norms (Madgavkar et al, 2020).
- The child-care burden of teaching at home during lockdowns is more likely to be placed on and performed by women (Alon et al, 2020).
- The ‘1.5 earner model’ describes the common situation where parents of young children earning sixty-thousand-dollar salaries or less each will lose part of their income, being as high as 90% or more, on childcare, taxation and government payments if the secondary earner were to work on the fourth day of the working week. This secondary earner is often the mother due to gender norms, disincentivizing her from working and increasing domestic responsibilities (Wood, Griffiths & Emslie, 2020). As more women continue to lose economic stability from the recession impacts outlined prior, this situation is likely to become more common following the end of the government’s childcare subsidies.

The COVID-19 pandemic has deteriorated conditions for women due to an increased threat of violence toward women. Firstly, the direct implication of the pandemic resulting in more people being encouraged and mandated to isolate to reduce the spread has placed women at a larger risk of domestic violence. Of women who have been reported physical or sexual violence since the beginnings of the pandemic in Australia, 66% are experiencing it for the first time, or it has escalated. Moreover, out of women who have reported coercive control, more than half have reported the acts began or escalated during the pandemic. In addition to this, violence toward women increases during times of economic downturn (WGEA, 2020). Lastly, the ability of women’s shelters to have capacity, or sufficient financial support, is less likely due to the nature of social distancing and the conditions of the economy (United Nations, 2020). Each of these impacts culminates to significantly worse conditions for women currently experiencing violence and increased risk to women who have yet to experience it. 

Evidently, the gendered impact of COVID-19 has been widely studied and addressed by the media and academia, specifically considering the short period of time the pandemic has been investigated for and lockdowns have been implemented. The problem has also been addressed and recognised at a global level by organisations such as the United Nations (United Nations, 2020) and the OECD (OECD, 2020). Despite this, there is a lack of government response, with many policies, as mentioned previously, increasing or ignoring the problem.
